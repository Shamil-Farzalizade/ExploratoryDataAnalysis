```python
x_train
```

```python
x_test
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor(random_state=42)
```

```python
model.fit(x,y)
```

```python
predictions= model.predict(x)
print(predictions)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
y_pred= model.predict(x_test)
mse= mean_squared_error(y_test, y_pred)
print(mse)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor_rf= RandomForestRegressor()
regressor_rf.fit(x_train, y_train)
```

```python
y_pred= regressor_rf.predict(x_test)
```

```python
mse= mean_squared_error(y_test, y_pred)
print(mse)
```

```python
att= df[df['Attrition']==1]
avg= att.groupby('Gender')['JobSatisfaction'].mean()
plt.figure(figsize=(8,8))
plt.pie(avg,labels= avg.index)
plt.title("Job Satisfaction for Attrition by Gender")
plt.show()
```

```python
att= df[df['Attrition']==1]
avg= att.groupby('Gender')['MonthlyRate'].mean()
plt.figure(figsize=(8,8))
plt.pie(avg,labels= avg.index)
plt.title("MonthlyRate for Attrition by Gender")
plt.show()
```

Generally we observed that Attrition_No was more with Current manager as well as with years in current role. As we analyzed the data, attrition rate was more for male rather than for females. Regarding Education Attrition rate was relatively high in 1. When we compared monthly income we noticed that it was more for females than for males. With regards to job satisfaction, by looking at pie chart we observed that males were relatively more satisfied with their job than females. Recommendations would be that maybe in order to keep a balance at work, monthly salaries for males should be changed along with job satisfaction should be altered for females. All these changes might have an impact on Attrition rate and thus Attrition rate could decrease due to these changes.
